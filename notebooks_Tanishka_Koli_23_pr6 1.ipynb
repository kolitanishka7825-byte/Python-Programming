```python
import sys

def main():
    # DECLARE cartA <- EMPTY SET
    cart_a = set()
    # DECLARE cartB <- EMPTY SET
    cart_b = set()

    while True:
        print("\n--- SHOPPING CART MENU ---")
        print("1. Add Products to Cart A")
        print("2. View Cart A Details (Items, Total, Max, Min)")
        print("3. Search for Product in Cart A")
        print("4. Remove Product from Cart A")
        print("5. View Premium Products (Limit Filter)")
        print("6. Add Products to Cart B")
        print("7. Perform Set Operations (Union & Intersection)")
        print("8. View Cart A in Sorted Order")
        print("9. Exit")
        
        choice = input("Enter your choice (1-9): ")
        print("-" * 30)

        # CHOICE 1: Add to Cart A
        if choice == '1':
            try:
                n = int(input("Enter number of products to add to Cart A: "))
                for i in range(n):
                    value = int(input(f"Enter Product ID {i+1}: "))
                    cart_a.add(value)  # ADD value TO cartA
                print("Products added successfully.")
            except ValueError:
                print("Invalid input! Please enter numbers only.")

        # CHOICE 2: View Cart A Details
        elif choice == '2':
            if not cart_a:
                print("Cart A is empty.")
            else:
                print("Products in Cart A:")
                for item in cart_a:
                    print(item)
                
                # SIZE, MAX, MIN
                print(f"Total Unique Products in Cart A: {len(cart_a)}")
                print(f"Highest Product ID: {max(cart_a)}")
                print(f"Lowest Product ID: {min(cart_a)}")

        # CHOICE 3: Search
        elif choice == '3':
            try:
                search_value = int(input("Enter Product ID to search: "))
                if search_value in cart_a:
                    print("Product found in cart")
                else:
                    print("Product not found")
            except ValueError:
                print("Invalid input.")

        # CHOICE 4: Remove
        elif choice == '4':
            try:
                remove_value = int(input("Enter Product ID to remove: "))
                if remove_value in cart_a:
                    cart_a.remove(remove_value)
                    print("Product removed")
                else:
                    print("Product not found")
            except ValueError:
                print("Invalid input.")

        # CHOICE 5: Filter Premium Products
        elif choice == '5':
            try:
                limit = int(input("Enter premium product ID limit: "))
                print("Premium Products:")
                found = False
                for item in cart_a:
                    if item >= limit:
                        print(item)
                        found = True
                if not found:
                    print("No premium products found above that limit.")
            except ValueError:
                print("Invalid input.")

        # CHOICE 6: Add to Cart B
        elif choice == '6':
            try:
                n = int(input("Enter number of products for Cart B: "))
                for i in range(n):
                    value = int(input(f"Enter Product ID {i+1}: "))
                    cart_b.add(value)
                print("Products added to Cart B.")
            except ValueError:
                print("Invalid input.")

        # CHOICE 7: Union and Intersection
        elif choice == '7':
            # unionCart <- cartA UNION cartB
            union_cart = cart_a.union(cart_b) # Or cart_a | cart_b
            
            # intersectionCart <- cartA INTERSECTION cartB
            intersection_cart = cart_a.intersection(cart_b) # Or cart_a & cart_b

            print(f"Union of Cart A and Cart B (All items): {union_cart}")
            print(f"Intersection of Cart A and Cart B (Common items): {intersection_cart}")

        # CHOICE 8: Sort
        elif choice == '8':
            if not cart_a:
                print("Cart A is empty.")
            else:
                print("Cart A in Sorted Order:")
                print(sorted(cart_a))

        # CHOICE 9: Exit
        elif choice == '9':
            print("Exiting program.")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 9.")

if __name__ == "__main__":
    main()
```
