```python
# student_marks = []
pass_mark = 40
distinction_mark = 75

while True:
    print("\n--- MENU ---")
    print("1. Add Student Mark")
    print("2. Display All Marks")
    print("3. Total and Average Marks")
    print("4. Highest and Lowest Marks")
    print("5. Passed and Failed Students")
    print("6. Distinction Students")
    print("7. Remove Duplicate Marks")
    print("8. Apply Grace Marks")
    print("9. Search a Mark")
    print("10. Sort Marks")
    print("11. Remove a Student Record")
    print("12. Find Index of a Mark")
    print("0. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        mark = int(input("Enter mark: "))
        student_marks.append(mark)
        print("Mark added successfully")

    elif choice == 2:
        print(student_marks)

    elif choice == 3:
        if student_marks:
            total = sum(student_marks)
            average = total / len(student_marks)
            print("Total:", total)
            print("Average:", average)
        else:
            print("No records available")

    elif choice == 4:
        if student_marks:
            print("Highest:", max(student_marks))
            print("Lowest:", min(student_marks))
        else:
            print("No records available")

    elif choice == 5:
        print("Passed Students:")
        for m in student_marks:
            if m >= pass_mark:
                print(m)

        print("Failed Students:")
        for m in student_marks:
            if m < pass_mark:
                print(m)

    elif choice == 6:
        print("Distinction Students:")
        for m in student_marks:
            if m >= distinction_mark:
                print(m)

    elif choice == 7:
        student_marks = list(set(student_marks))
        print("Duplicate marks removed")

    elif choice == 8:
        for i in range(len(student_marks)):
            if student_marks[i] < pass_mark:
                student_marks[i] = pass_mark
        print("Grace marks applied")

    elif choice == 9:
        search_mark = int(input("Enter mark to search: "))
        if search_mark in student_marks:
            print("Mark found")
        else:
            print("Mark not found")

    elif choice == 10:
        sorted_marks = sorted(student_marks)
        print("Ascending:", sorted_marks)
        print("Descending:", sorted_marks[::-1])

    elif choice == 11:
        mark_to_remove = int(input("Enter mark to remove: "))
        if mark_to_remove in student_marks:
            student_marks.remove(mark_to_remove)
            print("Student record removed")
        else:
            print("Mark not found")

    elif choice == 12:
        search_mark = int(input("Enter mark: "))
        if search_mark in student_marks:
            print("Index position:", student_marks.index(search_mark))
        else:
            print("Mark not found")

    elif choice == 0:
        print("Exiting Program")
        break

    else:
        print("Invalid choice")
```
