```python
books = ()
 
while True:
    print("\n===== LIBRARY INVENTORY SYSTEM =====")
    print("1. Enter book values")
    print("2. Display all book records")
    print("3. Total books and average value")
    print("4. Highest and lowest book value")
    print("5. Display premium books")
    print("6. Search for a book value")
    print("7. Count occurrence of a book value")
    print("8. Remove duplicate records")
    print("9. Display books in reverse order")
    print("10. Find index of a book value")
    print("11. Exit")
 
    choice = int(input("Enter your choice: "))
 
    # 1. Enter book data
    if choice == 1:
        n = int(input("How many books? "))
        temp = []
        for i in range(n):
            value = int(input(f"Enter book value {i+1}: "))
            temp.append(value)
        books = tuple(temp)
        print("Book data stored successfully.")
 
    # 2. Display all records
    elif choice == 2:
        print("Book Records:", books)
 
    # 3. Total books and average value
    elif choice == 3:
        if len(books) == 0:
            print("No books available.")
        else:
            total = sum(books)
            avg = total / len(books)
            print("Total books:", len(books))
            print("Average value:", avg)
 
    # 4. Highest and lowest value
    elif choice == 4:
        if len(books) == 0:
            print("No books available.")
        else:
            print("Highest value:", max(books))
            print("Lowest value:", min(books))
 
    # 5. Premium books
    elif choice == 5:
        limit = int(input("Enter premium limit: "))
        premium = [b for b in books if b >= limit]
        print("Premium books:", tuple(premium))
 
    # 6. Search a book value
    elif choice == 6:
        key = int(input("Enter value to search: "))
        if key in books:
            print("Book value found.")
        else:
            print("Book value not found.")
 
    # 7. Count occurrence
    elif choice == 7:
        key = int(input("Enter value to count: "))
        print("Occurrences:", books.count(key))
 
    # 8. Remove duplicates
    elif choice == 8:
        books = tuple(set(books))
        print("Duplicates removed.")
        print("Updated records:", books)
 
    # 9. Reverse order
    elif choice == 9:
        print("Reversed tuple:", books[::-1])
 
    # 10. Find index
    elif choice == 10:
        key = int(input("Enter value to find index: "))
        if key in books:
            print("Index:", books.index(key))
        else:
            print("Value not found.")
 
    # 11. Exit
    elif choice == 11:
        print("Exiting Library Inventory System.")
        break
 
    else:
        print("Invalid choice. Try again.")
```
