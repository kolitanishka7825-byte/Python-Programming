```python
# Bank Account Management System with Selection Structure
bankAccounts = {}
n = int(input("Enter number of accounts to create: "))
for i in range(n):
    accNo = input("Enter Account Number: ")
    balance = float(input("Enter Account Balance: "))
    bankAccounts[accNo] = balance
    
while True:
    print("\n--- MENU ---")
    print("1. Display all accounts")
    print("2. Total, Average, Highest, Lowest balance")
    print("3. Show high-value accounts")
    print("4. Search account")
    print("5. Deposit / Withdraw")
    print("6. Close account")
    print("7. Display sorted accounts")
    print("8. Exit")
    choice = int(input("Enter your choice: "))

    # Selection starts here
    if choice == 1:
        for accNo, balance in bankAccounts.items():
            print("Account:", accNo, "Balance:", balance)
    elif choice == 2:
        if bankAccounts:
            total = sum(bankAccounts.values())
            average = total / len(bankAccounts)
            highest = max(bankAccounts.values())
            lowest = min(bankAccounts.values())
            print("Total Balance:", total)
            print("Average Balance:", average)
            print("Highest Balance:", highest)
            print("Lowest Balance:", lowest)
        else:
            print("No accounts available")
    elif choice == 3:
        limit = float(input("Enter high-value balance limit: "))
        for accNo, balance in bankAccounts.items():
            if balance > limit:
                print(accNo, balance)
    elif choice == 4:
        searchAcc = input("Enter account number to search: ")
        if searchAcc in bankAccounts
            print("Account exists with balance:", bankAccounts[searchAcc])
        else:
            print("Account not found")
    elif choice == 5:
        accNo = input("Enter account number: ")
        if accNo in bankAccounts:
            amount = float(input("Enter amount (+deposit / -withdraw): "))
            bankAccounts[accNo] += amount
            print("Updated Balance:", bankAccounts[accNo])
        else:
            print("Account not found")
    elif choice == 6:
        accNo = input("Enter account number to close: ")
        if accNo in bankAccounts:
            del bankAccounts[accNo]
            print("Account removed")
        else:
            print("Account not found")
    elif choice == 7:
        for accNo in sorted(bankAccounts.keys()):
            print(accNo, bankAccounts[accNo])
    elif choice == 8:
        print("Exiting program")
        break
    else:
        print("Invalid choice")

 
```
