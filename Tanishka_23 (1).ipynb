3. Reverse a given number

```python
print("Tanishka , 23")
num = int(input("Enter a number:" ))
reverse = 0
while num > 0:
    reverse = reverse * 10 + num % 10
    num //=10

print("Reversed number: ", reverse)
```

4. Check whether a number is a palindrome using a loop


```python
print("Tanishka , 23")
num=int(input("Enter a number: "))
temp = num
reverse = 0

while temp > 0:
    reverse = reverse * 10 + temp % 10
    temp //= 10

if num == reverse:
    print("Palindrome")
else:
    print("Not a palindrome")
```

4. Check whether a given number is prime


```python
print("Tanishka , 23")
n=int(input("Enter a number:"))
count = 0

for i in range(1, n+1):
    if n % i ==0:
        count+=1
if count == 2:
    print("Prime number")
else:
    print("Not a prime number:")
```
